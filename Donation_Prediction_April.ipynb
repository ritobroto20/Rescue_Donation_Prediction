# Data Analysis

```python
plt.hist(df['TOTAL AMOUNT OF ECS'],bins=10,edgecolor='k',label="Histogram")
plt.xlabel("Rs.")
plt.ylabel("Frequency")
plt.title("TOTAL AMOUNT OF ECS", weight="bold", size=12)
plt.legend()
```

```python
plt.plot((df['MONTH'].iloc[:-1]),df['AMOUNT'].iloc[:-1],'*-',label="Datapoints")
plt.legend()
plt.ylim(300000,800000)
plt.title("Amount of Money received in the month", weight="bold")
plt.ylabel("Rs.")
```

```python
anount_dervtv= (df['AMOUNT'].iloc[1:12]-df['AMOUNT'].iloc[:11])
```

```python
amount_dervtv= (df['AMOUNT'].iloc[1:12].to_numpy() - df['AMOUNT'].iloc[:11].to_numpy())/df['AMOUNT'].iloc[:11].to_numpy()
amount_dervtv

plt.plot((df['MONTH'].iloc[1:-1]),amount_dervtv*100,'*-',label="Datapoints")
plt.axhline(y = (amount_dervtv*100).mean(), color = 'r', linestyle = '--', label="Mean growth")
plt.title("%age Growth per month", weight= 'bold')
plt.ylabel("%age")
plt.legend()
```

```python
(amount_dervtv*100).mean()
```

```python
(amount_dervtv*100).std()
```

```python
plt.figure(figsize=(5,6))
plt.errorbar(y=4.78,x=0, yerr = 3.09, fmt='-o', ecolor='k', color='r', elinewidth=2.5, capsize=10,label=f"Standard Deviation of Growth = 3.1")
plt.axhline(y = 4.78, color = 'r', linestyle = '--', label=f"Mean growth = 4.8")
plt.legend()
plt.ylim(0,10)
plt.title("Growth rate error bar", weight='bold')
plt.ylabel("%age growth")
plt.grid()
```
